```python
#Preparing data with Pandas

import pandas as pd
```

```python
#Building Model with Scikit Learn

from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('C:\My Projects\CODING\Python 3\Machine Learning\titanic.csv')
df['male'] = df['Sex'] == 'male'
X = df[['Pclass', 'male', 'Age', 'Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Fare']].values
y = df['Survived'].values


#Intiation of Class 
model = LogisticRegression()

#The fit method is used for building the model. It takes two arguments: X (the features as a 2d numpy array) 
#and y (the target as a 1d numpy array).
model.fit(X, y)

#Let’s create an array that has the predicted y values.
#Now we create an array of boolean values of whether or not our model predicted each passenger correctly.
y_pred = model.predict(X)

#To get the number of these that are true, we can use the NumPy sum method
#This means that of the 887 datapoints, the model makes the correct prediction for 714 of them.
print((y == y_pred).sum()) 

#how good our model is by counting the number of datapoints it predicts correctly. 
#This is called the accuracy score.

print(model.score(X, y)) 

```
